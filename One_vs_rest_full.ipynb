## Data processing

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.multiclass import OneVsRestClassifier
from sklearn.svm import SVC
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.metrics import classification_report, accuracy_score
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
```

```python
# Đọc dữ liệu
merged_data = pd.read_csv('merged_data.csv')

merged_data.info()
# Display the first few rows of the dataset
merged_data.head()
```

```python
# Delete all columns (genes) containing missing values
df = merged_data

# Check for missing values in the dataset
missing_values = df.isnull().sum()
missing_values = missing_values[missing_values > 0]
missing_values
missing_rows = df.isnull().any(axis=1)
df = df.drop(columns=missing_values.index)

#if 'Patient ID' in df.columns:
#    df = df.drop(columns=['Patient ID'])
df.head(5)
```

```python
# Encoding 'Subtype' column
encoder = LabelEncoder()
df['Subtype_encoded'] = encoder.fit_transform(df['Subtype']) + 1

# Assuming df has the 'Subtype_encoded' and feature columns
X = df.drop(['Subtype', 'Patient ID', 'Subtype_encoded'], axis=1)
y = df['Subtype_encoded']
df.head(5)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25)
```

```python
import pandas as pd

df = pd.DataFrame(X_train)
df.head()  # Now you can use the head() method
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import mutual_info_classif

# Select the best features
bestfeatures = SelectKBest(score_func=mutual_info_classif, k='all')
fit = bestfeatures.fit(X_train, y_train)
dfscores = pd.DataFrame(fit.scores_)
dfcolumns = pd.DataFrame(X_train.columns)

# concat two dataframes for better visualization
featureScores = pd.concat([dfcolumns, dfscores], axis=1)
featureScores.columns = ['Features', 'Score']
```

```python
featureScores.sort_values('Score', ascending=False)[0:30]
```

```python
# Select optimal features
X_optimal = X_train[featureScores.sort_values('Score', ascending=False).Features[0:500]]
X_tst_optimal = X_test[featureScores.sort_values('Score', ascending=False).Features[0:500]]
X_optimal.shape, X_tst_optimal.shape
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
# Cross-validation
clf = LogisticRegression(max_iter=200)
print(cross_val_score(clf, X_optimal, y_train, cv=10))
```

## 3.Gradient Boosting (One-vs-Rest)

```python
import numpy as np
import pandas as pd
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import classification_report
from sklearn.preprocessing import label_binarize

# Assume X_train, X_test, y_train, y_test are already defined as pandas DataFrames/Series

# Initial Gradient Boosting (One-vs-Rest) to get feature importances
ovr_classifier = OneVsRestClassifier(GradientBoostingClassifier())
ovr_classifier.fit(X_train, y_train)

# Predict the test set
y_pred = ovr_classifier.predict(X_test)

# Print classification report
print("Initial Gradient Boosting (OvR):")
print(classification_report(y_test, y_pred))

# Define the number of top features to select for each experiment
num_features_list = [20, 10, 5]

# Loop through each feature set size and evaluate model performance
for n in num_features_list:
    print(f"\nEvaluating with top {n} important features per class...")

    # Dictionaries to hold per-class selected features and classifiers
    class_feature_indices = {}
    class_classifiers = {}

    # For each class, select top n features and train classifier
    classes = ovr_classifier.classes_
    for idx, class_label in enumerate(classes):
        # Get the classifier for this class from the initial model
        classifier = ovr_classifier.estimators_[idx]
        importances = classifier.feature_importances_
        # Get indices of top n features for this class
        indices = np.argsort(importances)[::-1][:n]
        class_feature_indices[class_label] = indices

        # Select features for this class
        X_train_class = X_train.iloc[:, indices]
        # Binarize the labels for this class (One-vs-Rest)
        y_train_class = (y_train == class_label).astype(int)

        # Train classifier for this class using the selected features
        clf = GradientBoostingClassifier()
        clf.fit(X_train_class, y_train_class)
        class_classifiers[class_label] = clf

    # Now, predict on test data
    # Prepare an array to hold the decision scores
    test_scores = np.zeros((X_test.shape[0], len(classes)))

    for idx, class_label in enumerate(classes):
        indices = class_feature_indices[class_label]
        X_test_class = X_test.iloc[:, indices]
        clf = class_classifiers[class_label]

        # Get decision function or probabilities
        if hasattr(clf, "decision_function"):
            scores = clf.decision_function(X_test_class)
        else:
            scores = clf.predict_proba(X_test_class)[:, 1]
        test_scores[:, idx] = scores

    # Predict the class with the highest score
    y_pred = classes[np.argmax(test_scores, axis=1)]

    # Print the classification report
    print(f"\nGradient Boosting (OvR) - Retrained with Top {n} Features per Class:")
    print(classification_report(y_test, y_pred))
```

## 9.XGBoost & Light GBM with X and Y optimal

```python
from sklearn.metrics import classification_report
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from sklearn.multiclass import OneVsRestClassifier
from sklearn.model_selection import RandomizedSearchCV

# Define the models with GPU support
xgb = OneVsRestClassifier(XGBClassifier(tree_method='gpu_hist', predictor='gpu_predictor'))
lgbm = OneVsRestClassifier(LGBMClassifier(device='gpu'))

# Define the hyperparameters for RandomizedSearchCV
params_xgb = {
    'estimator__n_estimators': [50, 100, 200],
    'estimator__learning_rate': [0.01, 0.1, 0.2],
    'estimator__max_depth': [None, 5, 10, 15]
}

params_lgbm = {
    'estimator__n_estimators': [50, 100, 200],
    'estimator__learning_rate': [0.01, 0.1, 0.2],
    'estimator__max_depth': [None, 5, 10, 15]
}

# Perform hyperparameter optimization
random_search_xgb = RandomizedSearchCV(xgb, params_xgb, cv=3, n_jobs=-1, verbose=1)
random_search_xgb.fit(X_optimal, y_train - 1)

random_search_lgbm = RandomizedSearchCV(lgbm, params_lgbm, cv=3, n_jobs=-1, verbose=1)
random_search_lgbm.fit(X_optimal, y_train - 1)

# Print the best parameters
print("Best parameters for XGBoost with OvR: ", random_search_xgb.best_params_)
print("Best parameters for LightGBM with OvR: ", random_search_lgbm.best_params_)

# Evaluate on the test set
print("Test set score for XGBoost with OvR: ", random_search_xgb.score(X_tst_optimal, y_test - 1))
print("Test set score for LightGBM with OvR: ", random_search_lgbm.score(X_tst_optimal, y_test - 1))

# Predict the test set results
y_pred_xgb = random_search_xgb.predict(X_tst_optimal)
y_pred_lgbm = random_search_lgbm.predict(X_tst_optimal)

# Print the classification report
print("Classification report for XGBoost with OvR: ")
print(classification_report(y_test - 1, y_pred_xgb))

print("Classification report for LightGBM with OvR: ")
print(classification_report(y_test - 1, y_pred_lgbm))
```
